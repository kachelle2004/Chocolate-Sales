```python
sorted_country = df['Country'].value_counts().sort_values(ascending=False)
sorted_country.plot(kind='bar', figsize=(10,5))
plt.title('Number of Sales by Country') 
plt.xlabel('Country')
plt.ylabel('Number of Sales')   
plt.show()
```

```python
product_sales = df['Product'].value_counts().sort_values(ascending=False)
product_sales.plot(kind='bar', figsize=(10,5))
plt.title('Number of Sales by Product')
plt.xlabel('Product')
plt.ylabel('Number of Sales')
plt.show()
```

```python
total_sales_by_person = df.groupby('Sales Person')['Amount'].sum()
plt.bar(total_sales_by_person.index, total_sales_by_person.values) 
plt.title('Total Sales Amount by Sales Person')
plt.xlabel('Sales Person')
plt.ylabel('Total Sales Amount')
plt.show()

```
